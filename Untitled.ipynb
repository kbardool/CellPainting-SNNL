```python
model = Autoencoder(input_shape=784, code_dim=128)
```

```python
model = model.to(device)
```

```python
optimizer = torch.optim.Adam(params=model.parameters(), lr=1e-3)
```

```python
epochs = 50
```

```python
def loss(model, reconstructions, original, labels, epoch, optimizer):
    
    optimizer.zero_grad()
    
    bce_loss = torch.nn.BCEWithLogitsLoss()(reconstructions, original)
    
    activations = {}
    
    for index, layer in enumerate(model.layers):
        if index == 0:
            activations[index] = layer(original)
        else:
            activations[index] = layer(activations[index - 1])
    
    layers_snnl = []
    
    temperature = 1.0 / ((1.0 + epoch)**0.55)
    
    for key, value in activations.items():
        layer_snnl = SNNL(
            features=value,
            labels=labels,
            temperature=temperature,
        )
        layers_snnl.append(layer_snnl)
    
    del activations
    
#     snn_loss = sum(layers_snnl)
    snn_loss = torch.min(torch.Tensor(layers_snnl))
    
    train_loss = [bce_loss, (100. * snn_loss)]
    train_loss = sum(train_loss)
    
    train_loss.backward()
    optimizer.step()
    
    return train_loss, snn_loss, bce_loss
```

```python
train_loss = []
train_recon_loss = []
train_snn_loss = []

for epoch in range(epochs):
    epoch_loss = []
    epoch_recon_loss = []
    epoch_snn_loss = []
    for batch_features, batch_labels in train_loader:
        batch_features = batch_features.view(-1, 784).to(device)
        batch_labels = batch_labels.float()
        
#         optimizer.zero_grad()
        outputs = model(batch_features)
#         step_loss = torch.nn.MSELoss()(outputs, batch_features)
#         step_loss.backward()
#         optimizer.step()
        step_loss, snn_loss, recon_loss = loss(model, outputs, batch_features, batch_labels, epoch, optimizer)
               
        epoch_loss.append(step_loss.item())
        epoch_recon_loss.append(recon_loss)
        epoch_snn_loss.append(snn_loss)
        
    epoch_loss = torch.mean(torch.FloatTensor(epoch_loss))
    epoch_recon_loss = torch.mean(torch.FloatTensor(epoch_recon_loss))
    epoch_snn_loss = torch.mean(torch.FloatTensor(epoch_snn_loss))
    
    train_loss.append(epoch_loss)
    train_recon_loss.append(epoch_recon_loss)
    train_snn_loss.append(epoch_snn_loss)
    if (epoch + 1) % 10 == 0:
        print(f"epoch {epoch + 1}/{epochs} : epoch loss = {epoch_loss:.6f}")
        print(f"\trecon loss = {epoch_recon_loss}\t|\tsnn loss = {epoch_snn_loss}")
```

```python
def encode(train_features, test_features, components=3, seed=42):
    pca = PCA(n_components=components, random_state=seed)
    enc_train_features = pca.fit_transform(train_features)
    enc_test_features = pca.transform(test_features)
    return enc_train_features, enc_test_features
```
